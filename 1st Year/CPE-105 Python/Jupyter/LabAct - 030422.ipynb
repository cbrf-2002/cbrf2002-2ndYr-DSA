```python
from tkinter import *
from tkinter import messagebox as m

def addition():
    num1 = int(n1.get())
    num2 = int(n2.get())
    sums = num1+num2
    m.showinfo("Sum is: ", str(sums))
def subtraction():
    num1 = int(n1.get())
    num2 = int(n2.get())
    sums = num1-num2
    m.showinfo("Difference is: ", str(sums))
def multiplication():
    num1 = int(n1.get())
    num2 = int(n2.get())
    sums = num1*num2
    m.showinfo("Product is: ", str(sums))
def division():
    num1 = int(n1.get())
    num2 = int(n2.get())
    sums = num1/num2
    m.showinfo("Quotient is: ", str(sums))

root = Tk()
root.title("Charles Calculator | Form1")
root.geometry('400x200')

numtext1 = Label(root, text = "Num 1")
n1 = Entry(root, width = 30)
numtext2 = Label(root, text = "Num 2")
n2 = Entry(root, width = 30)

numtext1.grid(row=0, column=0)
numtext2.grid(row=1, column=0)
n1.grid(row=0, column=1)
n2.grid(row=1, column=1)

butsum = Button(root, text = "SUM", command = addition, width=20)
butdif = Button(root, text = "SUBTRACT", command = subtraction, width=20)
butmul = Button(root, text = "MULTIPLY", command = multiplication, width=20)
butdiv = Button(root, text = "DIVIDE", command = division, width=20)

butsum.grid(row=2, column=0)
butdif.grid(row=3, column=0)
butmul.grid(row=2, column=1)
butdiv.grid(row=3, column=1)

root.mainloop()
```

```python
from tkinter import *
from tkinter import messagebox as m

root = Tk()
root.title("Charles Fabian | Form 5")
root.geometry('300x100')

def vote():
    age = int(ent.get())
    if (age > 18):
        m.showinfo("WindowsApplication18", "You are eligible to vote")
    else:
        m.showinfo("WindowsApplication18", "Sorry, you are not eligible to vote")
        
enage = Label(root, text="Enter your age:")
ent = Entry(root)
butbut = Button(root, text = "Go", command = vote)

enage.pack()
ent.pack()
butbut.pack()

root.mainloop()
```
